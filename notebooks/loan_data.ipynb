```python
%matplotlib inline
import pandas as pd
import numpy as np
from urllib import request
```

```python
url = "https://www.nbp.pl/systemfinansowy/zrodlo-roz.xlsx"
```

Download data

```python
path = '../data/loans.xlsx'
# Download remote and save locally
data = request.urlretrieve(url, path)
```

Read data into dataframe

```python
df = pd.read_excel(path,sheet_name=0)
```

```python
df
```

```python
#pick row nr 3 as columns names
cols = df.iloc[1:2, 1:110].values
#slice the dataframe
df = df.iloc[108:115, 1:79]
#and change its column names
df.columns = cols[0]
df.rename(columns={'Kwartał, którego dotyczą pytania o stan (S)':'parameters'})
```

Lets transpose the table so the dates are in rows

```python
df = df.transpose()
df
```

Rename firts fow with columns name

```python
df.columns = df.iloc[0]
```

Remove first row

```python
df = df[1:]
```

```python
df
```

Lets work with data type in columns

```python
#reset index, quarter will be new column
df = df.reset_index()
df.columns = ['Quarter', 'Forcast', 'Changes_expenses', 'Alt_finansing', 'Change_economy', 'Change_rules', 'Change_criteria', 'Others']

#change type of all the columns with float type
cols = df.columns
df[cols[1:]] = df[cols[1:]].astype(float)
```

```python
#change date columns to datetime
dates = df['Quarter'].to_list()
new_dates = []
for date in dates:
    q,y = date.split('.')
    if q == 'I': q = 'Q1'
    if q == 'II': q = 'Q2'
    if q == 'III': q = 'Q3'
    if q == 'IV': q = 'Q4'
    new_dates.append(y + '-' + q)
new_dates

#update dates and change data type
df['Quarter'] = new_dates
df['Date'] = pd.to_datetime(df['Quarter'])

#delate Quarter column
df = df.drop('Quarter', axis=1)

df.dtypes
```
